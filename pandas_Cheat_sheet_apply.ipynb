# How to add two coloums at the same time?

Example 1: Adding columns with default or calculated values

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
}
df = pd.DataFrame(data)

# Adding two columns at the same time
df[['City', 'Country']] = [['New York', 'USA'], ['Los Angeles', 'USA'], ['Chicago', 'USA']]

print(df)
```

Example 2: Adding columns using calculations

```python
# Adding columns with calculated values
df[['Age in Months', 'Age in Days']] = pd.DataFrame({
    'Age in Months': df['Age'] * 12,
    'Age in Days': df['Age'] * 365
})

print(df)
```

4. Selecting Data

```python
# Selecting columns
df['Name']           # Accessing the 'Name' column (Series)
df[['Name', 'Age']]        # Accessing 'Name' and 'Age' columns (DataFrame)

# Selecting rows
df.iloc[0]                  # First row by index
# Assuming 'Age' column has values > 10
df.loc[df['Age'] > 10]      # Rows where 'Age' is greater than 10

# Slicing
df.iloc[0:3]                # First 3 rows
df.loc[0:2, ['Name']]       # Rows 0 to 2, and 'Name' column
```

5. Filtering and Sorting

```python
# Filtering
filtered_df = df[df['Age'] > 25]          # Rows where Age > 25

# Sorting
df.sort_values(by='Age', ascending=False)  # Sort by Age descending
```

6. Adding, Modifying, and Dropping Columns

```python
# Add new column
df['New_Col'] = df['Age'] * 2             # Based on existing column

# Modify column
df['Name'] = df['Name'].str.upper()       # Transform to uppercase

df
```

```python
# Drop columns
# df = df.drop(['Col_to_Drop'], axis=1)     # Original line causing the error
df = df.drop(['New_Col'], axis=1)  # Drop 'New_Col' or any existing column
# Make sure to replace 'New_Col' with the actual column name you want to drop.
```

8. Common DataFrame Operations

```python
# Apply function to an existing column, e.g., 'Age'
df['Age_doubled'] = df['Age'].apply(lambda x: x * 2)

# Value counts for 'Age' column
df['Age'].value_counts()

# Unique values in 'Age' column
df['Age'].unique()

# Renaming columns, e.g., rename 'Age' to 'Years'
df.rename(columns={'Age': 'Years'}, inplace=True)
```
