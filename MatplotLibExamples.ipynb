1️⃣ Univariate Analysis (One Variable)
Purpose: Understand the distribution of a single variable.

```python
# Load the dataset
df = pd.read_csv("sample_data.csv")

# Histogram (For Numerical Data)
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], bins=6, kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Box Plot (For Outlier Detection in Salary)
plt.figure(figsize=(5,4))
sns.boxplot(y=df['Salary'], color='green')
plt.title('Salary Distribution')
plt.ylabel('Salary')
plt.show()

# Bar Chart (For Categorical Data - Gender)
plt.figure(figsize=(5,3))
sns.countplot(x=df['Gender'], palette='coolwarm')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

2️⃣ Bivariate Analysis (Two Variables)
Purpose: Identify relationships between two variables.

```python
# Scatter Plot: Experience vs. Salary
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['Experience'], y=df['Salary'], hue=df['Gender'])
plt.title('Experience vs. Salary')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.show()

# Box Plot: Salary by Gender
plt.figure(figsize=(5,4))
sns.boxplot(x=df['Gender'], y=df['Salary'], palette='Set2')
plt.title('Salary Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Salary')
plt.show()

# Correlation Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm', linewidths=0.5) # Changed to only include numeric columns
plt.title("Correlation Heatmap")
plt.show()
```

3️⃣ Multivariate Analysis (Three or More Variables)
Purpose: Analyze multiple relationships at once.

```python
# Pair Plot (Shows relationship across all numerical columns)
sns.pairplot(df, hue='Gender', diag_kind='kde')
plt.show()

# 3D Scatter Plot (Using Matplotlib for 3D visualization)
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Age'], df['Experience'], df['Salary'], c='b', marker='o')

ax.set_xlabel('Age')
ax.set_ylabel('Experience')
ax.set_zlabel('Salary')
plt.title('3D Plot: Age vs Experience vs Salary')
plt.show()
```
